# where on earth is SST changing the fastest?

let's try to answer this question using NOAA's [Extended Reconstructed Sea Surface Temperature (ERSST) v5](https://www.ncdc.noaa.gov/data-access/marineocean-data/extended-reconstructed-sea-surface-temperature-ersst-v5)

We will start by looking at a few specific spots, then look at the whole globe


```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
url = 'http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/noaa.ersst.v5/sst.mnmean.nc'
```

```python
ds = xr.open_dataset(url, drop_variables=['time_bnds'])

print(ds.nbytes / 1e6, 'Mb')

ds
```

```python
# just grab the carriabian

carib = ds.sel(lon=slice(360-102,360-44), 
               lat=slice(35,2)).sel(time=slice('1900','2020' ))

carib
```

```python
carib.sst.sel(time='2020-09-01').plot()
```

```python
carib.sst.sel(lon = 290, lat = 18, method = 'nearest').plot()
```

```python
%%time
# remove the seasonal cycle:

gb = carib.groupby('time.month') 
carib_anom = gb - gb.mean(dim='time') 
carib_anom
```

```python
# plot a single point
# carib.sst.sel(lon = 290, lat = 18, method = 'nearest').plot()
carib_anom.sst.sel(lon = 290, lat = 18, method = 'nearest').plot()
```

```python
# average over space:

carib_anom.sst.mean( dim=['lon', 'lat']).plot()

# get a DECADE mean
# decade_mean = carib_anom.sst.rolling(time=10*12, center=True, min_periods=5).mean()
decade_mean = carib_anom.sst.coarsen(time=10*12, boundary='trim').mean()

decade_mean.mean( dim=['lon', 'lat']).plot()
```

```python
%%time
# do the same for the gulf of maine:


GOM = ds.sel(lon=slice(295,305), 
               lat=slice(50,44)).sel(time=slice('1900','2020' ))

gb = GOM.groupby('time.month') 
GOM_anom = gb - gb.mean(dim='time') 

# get a DECADE mean
# GOMdecade_mean = GOM_anom.sst.rolling(time=10*12, center=True, min_periods=5).mean()
GOMdecade_mean = GOM_anom.sst.coarsen(time=10*12, boundary='trim').mean()

```

```python
decade_mean.mean( dim=['lon', 'lat']).plot(label='carib')
GOMdecade_mean.mean( dim=['lon', 'lat']).plot(label='GOM')
plt.legend()
```

# fit a linear trend to the data and plot the trends

$ y = mx + b$

we can use python to fin $m$ and $b$

in our case $y$ is temperature and $x$ is time, so $m$ has units of degrees/time

```python


out = carib_anom.sst.polyfit(dim="time", deg=1, full=True)
out
```

```python
# plot the slope, or trend
out.polyfit_coefficients.sel(degree=1).plot()
```

```python
%%time
# try for the globe? 

globe120 = ds.sel(time=slice('1900','2020' )).coarsen(time=10*12, boundary='trim').mean()

out = globe120.sst.polyfit(dim="time", deg=1, full=True)

```

```python
out.polyfit_coefficients.sel(degree=1).plot( )
plt.ylim([40,60])
plt.xlim([275,345])
```

```python
plt.figure(figsize=(10,10))
plt.axes(projection = ccrs.Robinson())
out.polyfit_coefficients.sel(degree=1).plot( transform = ccrs.PlateCarree(), cbar_kwargs={'shrink':0.4})
```

```python
nstodecade = (1e9)*60*60*24*365*10

plt.figure(figsize=(10,10))
plt.axes(projection = ccrs.Robinson())

plt.pcolormesh(out.lon, out.lat, nstodecade*out.polyfit_coefficients.sel(degree=1),
              transform = ccrs.PlateCarree(), cmap='RdBu_r', vmin=-.2, vmax=.2)

plt.colorbar(shrink=0.4, label= '$^\circ$C / decade')
```
